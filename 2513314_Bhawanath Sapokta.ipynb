##Problem 2  South Asia Advanced Exploration

##Task 1  South Asia Subset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Define South Asian Countries
south_asian_countries = [
    "Afghanistan", "Bangladesh", "Bhutan", "India",
    "Maldives", "Nepal", "Pakistan", "Sri Lanka"
]
# Step 2: Filter dataset
filtered_df = df[df['Country name'].isin(south_asian_countries)].copy()
southasia_df = filtered_df.copy() # Assign to southasia_df for subsequent tasks
# Step 3: Save the filtered dataset
filtered_df.to_csv("south_asia_dataset.csv", index=False)
filtered_df.head()




```

```python
southasia_df.shape
```

##Task 2  Composite Score Ranking


```python
southasia_df['Composite Score'] = (
    0.40 * southasia_df['Log GDP per capita'] +
    0.30 * southasia_df['Social support'] +
    0.30 * southasia_df['Healthy life expectancy']
)

southasia_ranked = southasia_df.sort_values(by='Composite Score', ascending=False)
print("South Asia ranked by Composite Score:")
display(southasia_ranked[['Country name', 'Composite Score', 'score']])

# Top 5 horizontal bar chart
top5_comp = southasia_ranked.head(5)
plt.figure(figsize=(8,5))
sns.barplot(data=top5_comp, x='Composite Score', y='Country name', orient='h')
plt.title("Top 5 South Asian Countries by Composite Score")
plt.tight_layout()
plt.show()

# Compare Composite vs Original Score (scatter)
plt.figure(figsize=(6,5))
plt.scatter(southasia_df['Composite Score'], southasia_df['score'])
for _, row in southasia_df.iterrows():
    plt.text(row['Composite Score']+0.005, row['score']+0.005, row['Country name'], fontsize=8)
plt.xlabel("Composite Score")
plt.ylabel("Original Score")
plt.title("Composite Score vs Original Score (South Asia)")
plt.tight_layout()
plt.show()
```

##Task 3  Outlier Detection (1.5  IQR rule)

```python
# Outliers based on score
Q1_score = southasia_df['score'].quantile(0.25)
Q3_score = southasia_df['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score
lower_score = Q1_score - 1.5 * IQR_score
upper_score = Q3_score + 1.5 * IQR_score

# Outliers based on Log GDP per Capita
Q1_gdp = southasia_df['Log GDP per capita'].quantile(0.25)
Q3_gdp = southasia_df['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp
lower_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_gdp = Q3_gdp + 1.5 * IQR_gdp

outliers = southasia_df[
    (southasia_df['score'] < lower_score) | (southasia_df['score'] > upper_score) |
    (southasia_df['Log GDP per capita'] < lower_gdp) | (southasia_df['Log GDP per capita'] > upper_gdp)
]

print("Outlier countries in South Asia (by Score and GDP):")
display(outliers[['Country name', 'score', 'Log GDP per capita']])

# Scatter with outliers highlighted
plt.figure(figsize=(8,5))
plt.scatter(southasia_df['Log GDP per capita'], southasia_df['score'], label='Normal')
plt.scatter(outliers['Log GDP per capita'], outliers['score'], color='red', label='Outliers')
for _, row in outliers.iterrows():
    plt.text(row['Log GDP per capita']+0.01, row['score']+0.01, row['Country name'], fontsize=8)
plt.xlabel("Log GDP per Capita")
plt.ylabel("Score")
plt.title("South Asia: Score vs GDP (Outliers Highlighted)")
plt.legend()
plt.tight_layout()
plt.show()
```

##Task 4  Correlation & Trendlines


```python
metrics = ['Freedom to make life choices', 'Generosity']

for m in metrics:
    corr = southasia_df['score'].corr(southasia_df[m])
    print(f"Pearson correlation between Score and {m}:", corr)

    plt.figure(figsize=(6,4))
    sns.regplot(data=southasia_df, x=m, y='score')
    plt.title(f"Score vs {m} (South Asia)")
    plt.tight_layout()
    plt.show()
```

##Task 5  Gap Analysis (GDP  Score)

```python
southasia_df['GDP-Score Gap'] = southasia_df['Log GDP per capita'] - southasia_df['score']

# Rank ascending & descending
asc_gap = southasia_df.sort_values(by='GDP-Score Gap', ascending=True)
desc_gap = southasia_df.sort_values(by='GDP-Score Gap', ascending=False)

print("Ascending by GDP-Score Gap:")
display(asc_gap[['Country name', 'Log GDP per capita', 'score', 'GDP-Score Gap']])

print("\nDescending by GDP-Score Gap:")
display(desc_gap[['Country name', 'Log GDP per capita', 'score', 'GDP-Score Gap']])

# Top 3 positive & negative gaps
top3_pos = desc_gap.head(3)
top3_neg = asc_gap.head(3)

plt.figure(figsize=(8,5))
combined = pd.concat([top3_pos, top3_neg])
sns.barplot(data=combined, x='Country name', y='GDP-Score Gap')
plt.title("Top 3 Positive and Negative GDP-Score Gaps (South Asia)")
plt.xticks(rotation=90, ha='left')
plt.tight_layout()
plt.show()
```

##3) Metric Comparisons (Region-level averages)

```python
sa_means = southasia_df[['Log GDP per capita', 'Social support', 'Healthy life expectancy']].mean()
me_means = middleeast_df[['Log GDP per capita', 'Social support', 'Healthy life expectancy']].mean()

region_compare = pd.DataFrame({
    'Metric': ['Log GDP per capita', 'Social support', 'Healthy life expectancy'],
    'South Asia': sa_means.values,
    'Middle East': me_means.values
})

region_compare_melt = region_compare.melt(id_vars='Metric', var_name='Region', value_name='Value')

plt.figure(figsize=(8,5))
sns.barplot(data=region_compare_melt, x='Metric', y='Value', hue='Region')
plt.title("Key Metrics: South Asia vs Middle East")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

##4) Happiness Disparity (range + CV)

```python
def range_and_cv(series):
    _range = series.max() - series.min()
    _cv = series.std() / series.mean()
    return _range, _cv

sa_range, sa_cv = range_and_cv(southasia_df['score'])
me_range, me_cv = range_and_cv(middleeast_df['score'])

print("South Asia - Range:", sa_range, "CV:", sa_cv)
print("Middle East - Range:", me_range, "CV:", me_cv)
```

##5) Correlation Analysis (Score vs Freedom, Generosity)

```python
for region_name, region_df in [("South Asia", southasia_df), ("Middle East", middleeast_df)]:
    print(f"\nRegion: {region_name}")
    for col in ['Freedom to make life choices', 'Generosity']:
        corr = region_df['score'].corr(region_df[col])
        print(f"Correlation Score vs {col}: {corr}")

        plt.figure(figsize=(6,4))
        sns.regplot(data=region_df, x=col, y='score')
        plt.title(f"{region_name}: Score vs {col}")
        plt.tight_layout()
        plt.show()
```

##6) Outlier Detection (Score & GDP, both regions)

```python
def detect_outliers_iqr(df_region, col):
    Q1 = df_region[col].quantile(0.25)
    Q3 = df_region[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df_region[(df_region[col] < lower) | (df_region[col] > upper)]

sa_out = detect_outliers_iqr(southasia_df, 'score')
me_out = detect_outliers_iqr(middleeast_df, 'score')

print("South Asia Score Outliers:")
display(sa_out[['Country name', 'score', 'Log GDP per capita']])

print("\nMiddle East Score Outliers:")
display(me_out[['Country name', 'score', 'Log GDP per capita']])

# Scatter plots
def scatter_region_with_outliers(region_df, outliers_df, title):
    plt.figure(figsize=(8,5))
    plt.scatter(region_df['Log GDP per capita'], region_df['score'], label='Normal')
    if not outliers_df.empty:
        plt.scatter(outliers_df['Log GDP per capita'], outliers_df['score'], color='red', label='Outliers')
        for _, row in outliers_df.iterrows():
            plt.text(row['Log GDP per capita']+0.01, row['score']+0.01, row['Country name'], fontsize=8)
    plt.xlabel("Log GDP per Capita")
    plt.ylabel("Score")
    plt.title(title)
    plt.legend()
    plt.tight_layout()
    plt.show()

scatter_region_with_outliers(southasia_df, sa_out, "South Asia: Score vs GDP (Outliers)")
scatter_region_with_outliers(middleeast_df, me_out, "Middle East: Score vs GDP (Outliers)")
```
